## 岭回归 Ridge Regression
> 上面那种加入正则项的方式后新的目标就叫岭回归
![岭回归](images/岭回归.png)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.random.uniform(-3.0, 3.0, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x + 3 + np.random.normal(0, 1, size=100)
```

```python
plt.axis('on')
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

def PolynomialRegression(degree):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("lin_reg", LinearRegression())
    ])
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(666)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.metrics import mean_squared_error

poly_reg = PolynomialRegression(degree=20)
poly_reg.fit(X_train, y_train)

y_poly_predict = poly_reg.predict(X_test)
mean_squared_error(y_test, y_poly_predict) # 方差太大
```

```python
X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
y_plot = poly_reg.predict(X_plot)

plt.scatter(x, y)
plt.plot(X_plot[:,0], y_plot, color='r')
plt.axis([-3, 3, 0, 6])
plt.show()
```

```python
def plot_model(model):
    X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
    y_plot = model.predict(X_plot)

    plt.scatter(x, y)
    plt.plot(X_plot[:,0], y_plot, color='r')
    plt.axis([-3, 3, 0, 6])
    plt.show()

plot_model(poly_reg)
```

## 使用岭回归

```python
from sklearn.linear_model import Ridge # sklearn中的岭回归

def RidgeRegression(degree, alpha):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)), # 第一个参数
        ("std_scaler", StandardScaler()),
        ("ridge_reg", Ridge(alpha=alpha)) # 第二个参数
    ])
```

```python
ridge1_reg = RidgeRegression(degree=20, alpha=0.0001) # alpha就是岭回归超参数
ridge1_reg.fit(X_train, y_train)

y1_predict = ridge1_reg.predict(X_test)
mean_squared_error(y_test, y1_predict)
```

```python
plot_model(ridge1_reg)
```

```python
ridge2_reg = RidgeRegression(20, 1) # alhpa增大，曲线会更平滑
ridge2_reg.fit(X_train, y_train)

y2_predict = ridge2_reg.predict(X_test)
mean_squared_error(y_test, y2_predict)
```

```python
plot_model(ridge2_reg)
```

```python
ridge3_reg = RidgeRegression(20, 100) # alpha更大，曲线更加平滑
ridge3_reg.fit(X_train, y_train)

y3_predict = ridge3_reg.predict(X_test)
mean_squared_error(y_test, y3_predict)
```

```python
plot_model(ridge3_reg)
```

```python
ridge4_reg = RidgeRegression(20, 10000000) # alhpa太大也不好，最后我们的优化目标只剩正则化项了，这时所有的theta都等于0才是最优，拟合地曲线就成了一条直线了
ridge4_reg.fit(X_train, y_train)

y4_predict = ridge4_reg.predict(X_test)
mean_squared_error(y_test, y4_predict)
```

```python
plot_model(ridge4_reg)
```
