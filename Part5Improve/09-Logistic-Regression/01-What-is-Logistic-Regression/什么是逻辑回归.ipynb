## 逻辑回归解决地是`分类问题`
> 线性回归主要用来解决连续值预测的问题,而逻辑回归用来解决分类的问题,输出的属于某个类别的概率

回归方法怎么解决分类问题？

将样本的特征和样本发生的概率联系起来，概率是一个数，举例如下图
![逻辑回归举例](images/逻辑回归举例.png)
+ 逻辑回归既可以看做是回归算法，也可以看做是分类算法
+ 通常作为分类算法用，只可以解决二分类问题

## 传统的函数其值方位通常不在0~1之间，所以引入sigmoid函数

![Sigmoid函数把输出值范围限制在0到1之间](images/Sigmoid函数把输出值范围限制在0到1之间.png)

## sigmoid 函数使用演示

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(t):
    return 1. / (1. + np.exp(-t))
```

```python
x = np.linspace(-10, 10, 500)
plt.axis('on')
plt.plot(x, sigmoid(x))
plt.show() # 从下图可以看出y=sigmoid(x)被限制在[0, 1)之间
```
