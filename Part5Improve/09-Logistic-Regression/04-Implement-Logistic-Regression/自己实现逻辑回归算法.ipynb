# 自己实现的逻辑回归算法

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X[:5, :]
```

```python
y.shape
```

```python
y[:5]
```

```python
X = X[y<2, :2] # 选取X中对应地y小于2的部分的前两列(即前两个特征)
```

```python
y = y[y<2]
```

```python
X.shape
```

```python
y.shape
```

```python
plt.axis('on')
plt.scatter(X[y==0, 0], X[y==0, 1], color='red') # 用前两列即两个特征作为坐标轴的横纵坐标
plt.scatter(X[y==1, 0], X[y==1, 1], color='green')
plt.show()
```

## 使用我们自己实现地逻辑回归算法

```python
from playML.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, seed=666) # 默认按照测试集:训练集=8:2来划分地
```

```python
# 训练逻辑回归的模型
from playML.LogisticRegression import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test) # 预测正确率到了100%
```

```python
log_reg.predict_proba(X_test) # 每个元素对应的可能性,根据是否大于0.5决定是1还是0
```
