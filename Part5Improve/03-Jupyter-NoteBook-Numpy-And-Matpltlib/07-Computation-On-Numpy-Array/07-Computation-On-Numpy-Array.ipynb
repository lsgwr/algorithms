# numpy.array进行矩阵运算

给定一个向量，让想两种的每一个数乘以2 <br>
a = (0, 1, 2) <br> 
a*2 = (0, 2, 4) <br>

```python
n = 10
L = [i for i in range(n)]
```

```python
L
```

```python
2 * L  # 这种方式并不能达到效果
```

```python
A = []
for e in L:
    A.append(2*e)
A
```

```python
n = 1000000
L = [i for i in range(n)]
```

```python
%%time
A = []
for e in L:       # 用传统的for循环很慢
    A.append(2*e)
```

```python
%%time
A = [2*e for e in L] # 用生成表达式明显快很多
```

```python
import numpy as np
L = np.arange(n)
```

```python
%%time
A = np.array(2*e for e in L)  # 用numpy生成数组的速度明显快多了
```

```python
%%time
A = 2 * L  # numpy中的*2正是我们想要的结果，而且效果要快地多
```

```python
A
```

## Universla Functions(数字和矩阵之间的运算)

```python
x = np.arange(1, 16).reshape((3, 5))
x
```

```python
x + 1   # 矩阵加法
```

```python
x - 1 # 矩阵减法
```

```python
x * 2 # 乘法
```

```python
x / 2  # 浮点数除法
```

```python
x // 2 # 整数除法
```

```python
1 / x # 取倒数
```

```python
np.abs(x)  # 求绝对值
```

```python
np.sin(x) # 正弦函数
```

```python
np.cos(x) # 求余弦
```

```python
np.tan(x) # 余弦
```

```python
np.exp(x) # 求e的x次方
```

```python
np.power(3, x) # 求3的x次方， 等效于3**x
```

```python
3**x
```

```python
np.log(x) # 取对数值
```

```python
np.log2(x) #以2为底的对数
```

```python
np.log10(x) #以10为底的对数
```
