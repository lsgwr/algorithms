+ ### 数据访问

```python
x
```

```python
x[5]
```

```python
X[0][0] # 不推荐这么访问
```

```python
X[2, 2] # 推荐这样访问
```

```python
x[1:5] # 访问0~5的元素(不含5)
```

```python
x[:5] # 开头不写，默认从0开始访问
```

```python
x[5:] # 结尾不写，默认到最后
```

```python
x[::2] # 从头(第1个参数)访问到尾(第2个参数)，步长为2(第3个参数)
```

```python
x[::-1] # 从头(第一个参数)访问到尾(第二个参数)，步长为-1,相当于倒序访问
```

```python
X
```

```python
X[:2, :3] # 访问前两行(2表示行作为从开始位置到2，  3表示从开始位置到第3列)
```

```python
X[:2][:3] # 这样访问是不能正常访问前2行前3列地，访问地地是先取前2行，再从取地结果中取前3行，因为只有两行，所以前两行内容全部打印
```

```python
X[:2, ::2] # 访问前两行，间隔为2
```

```python
X[::-1, ::-1] # 行和列都倒着访问，相当于转置了
```

```python
X[0] # 访问第一行
```

```python
X[0 ,:] # 访问第一行，写上第二个参数
```

```python
X[:, 0] # 只取第一列
```

```python
subX = X[:2, :3] #取两行3列作为子矩阵
subX
```

```python
subX[0][0] = 100 # 取子矩阵，通过应用取地，所以子矩阵修改会影响父矩阵
subX
```

```python
X # 子矩阵修改影响父矩阵
```

```python
X[0][0] = 0 # 修改父矩阵也影响子矩阵
X
```

```python
subX
```

```python
subX2 = X[:2, :3].copy()  # 获取真正的子矩阵，并且父子矩阵修改互不影响
subX2
```

```python
subX2[0][0] = 100 
subX2
```

```python
X  # copy后生成的子矩阵修改不影响父矩阵
```

+ ### Reshape 重新调整数组的维度

```python
x.shape
```

```python
x.ndim
```

```python
A = x.reshape(2, 5)   # 把10个元素的一维数组，变成2行5列的二维数组
A
```

```python
B = x.reshape(1, 10) # 调整成1行10列的二维数组(没错，就是二维数组)，注意看中括号的个数
B
```

```python
B.ndim  # 维度是2，说明是二维数组，印证了上面的结论
```

```python
x.reshape(5 ,-1) # 改为10行x列(-1表示不在乎有多少列)的数组,5行是必须地
```

```python
x.reshape(-1 , 2) # 改为x行(-1表示不在乎有多少行)2列的数组，2列是必须地
```

```python
x.reshape(-1 , 5) # 改为x行(-1表示不在乎有多少行)5列的数组，5列是必须地，但是前提一定要正整除！！
```
